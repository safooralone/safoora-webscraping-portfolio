```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("http://www.imdb.com/chart/top/")
```

```python
movies=driver.find_elements(By.CSS_SELECTOR,"li.ipc-metadata-list-summary-item")
data=[]

```

```python
for x in movies:
    t=x.find_element(By.CSS_SELECTOR,"h3.ipc-title__text").text
    y=x.find_element(By.CSS_SELECTOR,"span.cli-title-metadata-item").text
    r=x.find_element(By.CSS_SELECTOR,"span.ipc-rating-star--rating").text
    print(t,y,r)
                     
                     
                     
```
